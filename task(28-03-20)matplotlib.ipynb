<pre>

</pre>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#x axis values
x = [1,2,3]
# y axis values
y = [2,4,1]
plt.plot(x,y, marker = '^', color = 'c')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
```

Display a bar chart of the popularity of programming Languages. <br>
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6,8.8,8,7.7, 6.7]
plt.bar(languages, popularity, width = 0.9, edgecolor = 'k', linestyle = '-.', color = 'c' )
```

```python
plt.barh(languages, popularity, edgecolor = 'k', linestyle = '-.', color = 'c',  orientation = 'horizontal')
```

Display a bar chart of the popularity of programming Languages. Use different color for each bar

```python
plt.bar(languages, popularity, color = 'bckymg',  edgecolor = 'k', linestyle = '-.' )
```

For the above data create a pie chart

```python
colors = ['magenta', 'gold', 'lightcoral', 'lightskyblue', 'yellowgreen', 'cyan']

plt.pie(popularity,explode = (0.06, 0,0,0, 0, 0), autopct = '%0.01f%%', shadow = True, colors = colors)

plt.tight_layout()
plt.legend(languages, bbox_to_anchor=(1, 1.05))
```

Draw a scatter plot  taking a random distribution in X and Y and plotted against each other<br>
use random.randn function

```python
x =np.random.randn(5,10)
y = np.random.randn(5,10)
plt.scatter(x,y, edgecolor = 'k', marker = 'X', cmap = 'twilight')
```

Draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other<br>
use random.randn function

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, color = 'gray', facecolors = 'none', s= 50)
```

draw a scatter plot using random distributions to generate circles of different sizes.

```python
x = np.random.randn(100)
y = np.random.rand(100)
colors = np.random.randn(100)
sizes = np.random.rand(100)
plt.scatter(x,y, c = colors, s = 100 * sizes, cmap = 'viridis', alpha = 0.5)
plt.colorbar()
```

draw a histogram owith your choice of data generated with numpy

```python
plt.figure(figsize=[10,8])

x = np.random.randn(1000)

plt.hist(x, bins = 5)
plt.grid(axis = 'y', alpha = 0.4)
plt.xlabel('Value',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.title('Histogram',fontsize=15)
```

```python
plt.figure(figsize=[10,8])

x = 0.3*np.random.randn(100)
y = 0.3*np.random.randn(50)

plt.hist([x,y], label = ['x', 'y'])
plt.grid(axis = 'y', alpha = 0.4)
plt.xlabel('Value',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.title('Histogram',fontsize=15)
plt.legend()
```
